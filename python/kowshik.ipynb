```python
class student:
    school_name='brilliant school'
    s_address='hyd'
    s_phone=1077
    s_details={}
    s_totalstudents={}
    no_students=0
    def __init__(self,st_name,phone,email,address,mark1,mark2,mark3):
        self.st_name=st_name
        self.phone=phone
        self.email=email
        self.address=address
        self.mark1=mark1
        self.mark2=mark2
        self.mark3=mark3
        self.no_of_students()
        self.rollno=100+self.no_students
        self.add_std(self.rollno,self)
        self.cal_total(self.mark1,self.mark2,self.mark3,self.rollno)
        
    @classmethod
    def add_std(cls,rollno,self):
        cls.s_details[rollno]=self

    @classmethod
    def no_of_students(cls):
        cls.no_students+=1
    
    @classmethod
    def cal_total(cls,mark1,mark2,mark3,roll):
        cls.s_totalstudents[roll]=mark1+mark2+mark3
    
    def modify_marks(self):
        n=int(input("selct 1.mark1 2.mark2 3.mark3 4.exit" ))
        match n:
            case 1:
                mark1=int(input("enter the new marks"))
                self.mark1=mark1
                self.cal_total(self.mark1,self.mark2,self.mark3,self.rollno)
            case 2:
                mark2=int(input("enter the new marks"))
                self.mark2=mark2
                self.cal_total(self.mark1,self.mark2,self.mark3,self.rollno)
            case 3:
                mark3 =int(input("enter the new marks"))
                self.mark3=mark3
                self.cal_total(self.mark1,self.mark2,self.mark3,self.rollno)
            case 4:
                return
        self.modify_marks()
    @classmethod
    def main(cls):
        m=int(input("enter 1.modify marks 2.display all students details 3.specific student details 4.changephone number 5.exit "))
        match m:
            case 1:
                roll=int(input("enter student roll no"))
                obj=cls.s_details[roll]
                obj.modify_marks()
            case 2:
                cls.display()
            case 3:
                roll=int(input("enter student roll no"))
                cls.specific_std(roll)
            case 4:
                cls.change_school_phone()
            case 5:
                return
    @classmethod
    def display(cls):  
        for i in cls.s_details:
           print(cls.s_details[i].st_name,
            cls.s_details[i].phone,
            cls.s_details[i].email,
            cls.s_details[i].rollno,
            cls.s_details[i].mark1,
            cls.s_details[i].mark2,
            cls.s_details[i].mark3,
            cls.s_details[i].address)
    @classmethod
    def specific_std(cls,roll):
        obj=cls.s_details[roll]
        print(obj.st_name,
            obj.phone,
            obj.email,
            obj.rollno,
            obj.mark1,
            obj.mark2,
            obj.mark3,
            obj.address)
    
    @classmethod
    def change_school_phone(cls):
        cls.s_phone=int(input('enter new phone number'))

    
khan=student('khan_sab',9123456784,'khan@gmail.com','vizag',77,87,97)
khan2=student('projj',9123456784,'khan@gmail.com','vizag',77,87,97)


```
